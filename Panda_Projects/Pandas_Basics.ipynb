### loc and iloc

```python
coffee.loc[0]     #coffee.loc[row,column]to view specific columns and rows of data we use loc or iloc
```

```python
coffee.loc[[0,1,4,5]]     #specific rows
```

```python
coffee.loc[0:8]     #slicing is possible in loc accesing
```

```python
coffee.loc[0:8,["Day","Units Sold"]]     # you can use combination of rows and columns acccesing
```

```python
coffee.iloc[:,[0,2]]      
#iloc is similar to loc but insted of using labels like day and units sold we use indexing in iloc
```

## Assigining Values

```python
coffee.loc[1,'Units Sold']=10
```

### Using at and iat

```python
coffee.at[0,"Units Sold"]      #at and iat can be used to access only that one single value
```

### Getting Columns

```python
coffee.Day         #works with non spaced column names
```

### Sort Values

```python
coffee.sort_values(["Units Sold"],ascending=False)
```

```python
coffee.sort_values(["Units Sold","Coffee Type"], ascending=False)      
#first it will sort by units sold and next it will again sort by lexographical/alphabeticall values look at the diff between previous and now sorting prev:10 8 12 6 4   but now the order is 8 10 12 6 9
```
